```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X=3 # assumed slope
a=np.arange(-2,2,0.15)
a=a.reshape(-1,1)
b=X*a+np.random.randn(*a.shape) # add noise

plt.plot(a, X*a, color ='b',LineWidth =2,label='true Line') #True relation
plt.plot(a,b,'X',color='r',MarkerSize=10,label='noisy data') # noisy relation


U,S,VT = np.linalg.svd(a,full_matrices=False)
#xtild=VT.T @ np.linalg.inv(np.diag(S))@U.T @ b # least square fit
#or we can use this also

xtild=np.linalg.pinv(a)@b # least square fit 

plt.plot(a,xtild * a,'--',color ='black',Linewidth=4,label='Regression line')

plt.xlabel('a')
plt.ylabel('b')

plt.grid(linestyle='--')
plt.legend()
plt.show()
```
